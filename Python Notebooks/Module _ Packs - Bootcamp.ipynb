**Creating a Package:**

1 - Create a Folder<br>
2 - Create a Subfolder<br>
3 - Create __init__.py file in MainFolder<br>
4 - Create __init__.py file in Subfolder<br>
5 - Create the main_script in MainFolder<br>
6 - Create the sub_script in SubFolder<br>
7 - To use that:<br>
      from Folder import main_script<br>
      from Folder.SubFolder import sub_script<br>
8 - To use especific function:<br>
      main_script.myfunc()<br>
