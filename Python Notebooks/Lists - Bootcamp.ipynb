```
#CURLY BRACES
#.append(new_item) ADD ITEM
list2 = [2, 5, 7, 1]
list1 = ["larry","darry","garry"]
list1.append("marry")
print(list1)

#.pop(index) REMOVES ITEM
list1.pop(1)
print(list1)

#.sort() ORDER THE LIST
list1.sort()
list2.sort()
print(list1, list2)

#.reverse()
list1.reverse()
list2.reverse()
print(list1, list2)
```
