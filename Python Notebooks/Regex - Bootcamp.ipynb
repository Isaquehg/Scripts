# Regex Patterns

* w matches alphanumeric characters, which means a-z, A-Z, and 0-9. It also matches the underscore, _, and the dash, -.
* d matches digits, which means 0-9.
* s matches whitespace characters, which include the tab, new line, carriage return, and space characters.
* S matches non-whitespace characters.
* . matches any character except the new line character n.


# Quantifiers
* '*'	Match zero or more times.
* '+'	Match one or more times.
* '?' Match zero or one time (plurals or not).
* {n} Match exactly n times.
* {n ,}	Match at least n times.
* {n , m}	{n , m}?	Match from n to m times.



```python
#finding phone numbers
import re
text = 'His phone number is 35_99999-1111'
phone = re.search(r'(\d{2})_(\d{5})-(\d{4})', text)
phone.group()#grouping the results
```

```python
#finding emails
text1 = 'my email is account@gmail.com'
pattern = re.compile(r'\w{1,}@\w{1,}.com')#some_name@some_mail.com
emails = re.search(pattern, text1)
emails.group()
```

```python
#excluding numbers
import re
phrase = "There are 5 numbers 75 inside 7 sentences. There are punctuations too!!"
pattern = r'[^\d!?. ]+'#using ^ to remove specific caracthers
final_phrase = re.findall(pattern, phrase)
" ".join(final_phrase)
```
