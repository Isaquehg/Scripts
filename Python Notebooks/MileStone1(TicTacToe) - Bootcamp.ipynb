```
#USER CHOICE test
def user_choice():
  choice = "wrong"
  acceptable_range = range(0,11)
  within_range = False

  #LOOP
  while choice.isdigit() == False or within_range == False:
    choice = input("enter a number 0 to 10: ")
    
    #Digit Check
    if choice.isdigit() == False:
      print("This is not a digit!")

    #Range Check
    if choice.isdigit() == True:
      if int(choice) in acceptable_range:
        within_range == True
        break
      else:
        print("Out of acceptable range!")

  return int(choice)
```
