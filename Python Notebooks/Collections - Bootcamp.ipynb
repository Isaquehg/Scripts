```python
#COUNTER
from collections import Counter 
#counter makes a dictionary from the items and tells how much there are!
#counter has the same dictionaries attributes!!!
#ex.: x = Counter(mylist)
#sum(x.values())

mylist = [1, "bla", "abc", 12, 15, 12, 1, 1, 1]

Counter(mylist)
```
