```python
import PyPDF2
```

```python
#Number of pages
f = open('ebook.pdf', 'rb')
reader = PyPDF2.PdfFileReader(f)
reader.numPages
```

```python
#Reading a page
first_page = reader.getPage(0)
text = first_page.extractText()
text
f.close()
```

```python
#Writing a page
f = open('ebook.pdf', 'rb')
reader = PyPDF2.PdfFileReader(f)
first_page = reader.getPage(0)
writer = PyPDF2.PdfFileWriter()
writer.addPage(first_page)
output = open('new_pdf.pdf', 'wb')
writer.write(output)
f.close()
```
