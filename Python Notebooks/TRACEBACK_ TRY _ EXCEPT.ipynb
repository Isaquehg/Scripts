```
x = "Hello Bob"
try:
  y = int (x)
except:
  y = -1
print(y)
a = "176"
try:
  b = int (a)
except:
  b = -2
print(b)
print ("Traceback Test!")
```

```
Larry = input("Enter a number: ")
try:
  Darry = int(Larry)
except:
  Darry = -1
if Darry > 0:
  print("Good Job!")
else:
  print("Not a Number...")
```
