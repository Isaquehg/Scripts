```python
def new_dec(original_func):
  
  def wrap_func():
    print("Before original funtion")

    original_func()

    print("After")

  return wrap_func
```

```python
@new_dec
def func_needs_dec():
  print("I want to be decorated!!")
```

```python
func_needs_dec()
```
