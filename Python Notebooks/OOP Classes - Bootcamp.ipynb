```
class Dog():
  
#The first function must have __init__
#And ALL the functions must have SELF inside them to get the correct reference
  def __init__(self,breed,name,spots):
    self.breed = breed
    self.name = name
    self.spots = spots
  
  def bark(self,number):
    print("WOOF!, WOOF!. My name is {} and i´m {} years old".format(self.name,number))
#Note that we have to
#use the self. before name to reference to the correct name!!
```

```
mydog = Dog("German Shephered","Sam",True)
mydog.breed
```

```
mydog.bark(10)
```

```
class Circle():
  pi = 3.14 
  #We can reference to the classes object by using
  #(classname.object)Circle.pi or (self.object)self.pi
  def __init__(self,radius=1):
    self.radius = radius
    self.area = radius * radius * Circle.pi
  def circunference(self):
    return self.radius * Circle.pi * 2
```

```
mycircle = Circle()
mycircle.circunference()
```
