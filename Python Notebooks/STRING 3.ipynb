```
#FIND
#When there´s no letter, the program outputs -1.

fruit = "apple"
position = fruit.find("e")
print(position)
blabla = fruit.find("w")
print(blabla)
```

```
#SEARCH & REPLACE:
#It replaces ALL the occurences!
greet = "Hello Bob"
nstr = greet.replace("Bob", "Jane")
print(nstr)
pstr = greet. replace("o", "x")
print(pstr)
```

```
#STRIPS
#STRIP, RightSTRIP, LeftSTRIP: Each of them remove a side and STRIP removes both.
greet = "  Hello  Bob  "
x = greet.lstrip()
print(x)
y = greet.rstrip()
print(y)
z = greet.strip()
print(z)
```

```
data = "From stephen.marquard@uct.ac.za Sat Jam  5 09:14:16 2008"
atpos = data.find("@")
print("Atpos:", atpos)
sppos = data.find(" ", atpos)
print("Sppos:", sppos)
host = data[atpos + 1: sppos]
print("Host:", host)
```
