```
my_string = ("abcdefgh")
print(my_string[::])#Take all the content.
print(my_string[::2])#jumping 2 in 2
print(my_string[::-1])#starting from the end

my_name = input("What´s your name? ")
print("Hello " + my_name)
```

```
#.FORMAT()
#Format insert string into {}. WE CAN INSERT THE INDEX INTO THE {0,exemple}

x = input("My favorite toys: ")

sep = x.split()
print("Your favorite toys are: {}".format(sep))
```

```
#formating
result = 100/700
#.format(variable) var:width.decimal numbers after
print("the result was {r:1.3}".format(r = result))
```
