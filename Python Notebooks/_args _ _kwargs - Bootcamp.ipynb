```
#Function to find 5% of a number sum

def myfunc(a,b):
  return sum((a,b)) * 0.05

myfunc(40,60) #5% of 100
```

```
#*ARGS
#The name mustn´t be ARGS. We can put any name like *abc...
def myfunc(*args):
  print(args)#*arg creates a TUPLE!
  return sum(args) * 0.05

#*args let us to add infinite arguments into the func!
myfunc(40,60,90,100)
```

```
#**KWARGS
def myfunc(**kwargs):
  print(kwargs)#NOTE that ** creates a DICTIONARY and * creates a tuple.

  if "fruit" in kwargs:
    print("My favorite fruit is {}".format(kwargs["fruit"]))
  else:
    print("There´s no fruit here!")

myfunc(fruit="orange")
```

```
#Bringing them together
def somefunc(*arg,**kwarg):
  print(arg)
  print(kwarg)
  print("I´d like {} {}".format(arg[0],kwarg["food"]))

somefunc(10,20,30,food="eggs",animal="dogs",auto="car")
```

```
#Creating a list with even number using random args
list1 = []
def myfunc(*args):
    for i in args:
        if i % 2 == 0:
            list1.append(i)
        else:
            pass
    return list1

myfunc(0,1,2,3,4,5,6,7)
```
