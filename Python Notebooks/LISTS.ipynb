```
#Here we can change a content of a list, but it´s not possible doing this for a string!!
number = [1, 54, 863, 75]
print(number)
number[2] = 22
print(number)
```

```
#RANGE AND LENGHT FUNCTIONS
#Range counts the number util the selected and returns it.
#Len counts the elements number in a list or string.
print(range(10))
videogames = ["PS4", "XBOX ONE", "PS5", "XBOX X"]
print(len(videogames))
print(range(len(videogames)))
```
