```
%%writefile mytext.txt
Hello this is a text file
second line
thrid line
```

```
myfile = open("/content/mytext.txt")

#WIN PATH: open("C:\\Users\\UserName\\Folder\\text.txt")

#OBSERVATION!!!!!!!
#If u want to read the file AGAIN, you have got to type 
#MYFILE.SEEK(0) to go back the reader cursor...

#Files methods: .read(), .readlines(), .close()
myfile.read()

myfile.read()#again
myfile.seek(0)#return the cursor to zero
myfile.readlines()#It reads the lines now from the zero position

myfile.close()#remember to do it always!
```

```
#ANOTHER WAY
#Using this way, you don´t need to close the file...
with open("/content/mytext.txt") as new_file:
  content = new_file.read()

print(content)
```

```
#MODES:
#r = read/ w = write or create new/ a = append/
#r+ = reading and writing/ w+ = writing and reading

#OVERWRITING THE EXISTING FILE
%%writefile mytext.txt
new first LINE
new second ONE
THRID ONE
```

```
#Appending another line
with open("/content/mytext.txt", mode="a") as f:
  f.write("\nFOUR LINE!")
```

```
#Reading the changed file:
with open("/content/mytext.txt") as new_file:
  content = new_file.read()
print(content)
```

```
#Writing a new file:
with open("randomfile.txt", mode="w") as rf:
  rf.write("I just created it!")
```

```
#Reading the last written file...
with open("randomfile.txt", mode="r") as rf:
  print(rf.read())
```
