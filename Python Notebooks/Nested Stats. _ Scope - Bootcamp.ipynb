```
#This is the priority order of Python´s scopes(LEGB Rules):
#L: local
#E: enclosing
#G: global
#B: built-in functions(len, max, range, etc.)
#GLOBAL
name = "THIS IS A GLOBAL STR"

def greet():
  
  #ENCLOSING
  name = "Sammy"

  def hello():
    #LOCAL
    name = "i´m local"
    print("Hello", name)

  hello()

#calling everything...
greet()
```
