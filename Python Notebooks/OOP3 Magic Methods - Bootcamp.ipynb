```
#This methods are built-in functions for classes like print and len for things likes lists.
#When I print a class witch has no __str__, will apear just the LOCATION of that!
 #This __str__ will return the NAME I wanna appear when I PRINT OUT the Class!

class Book():
  def __init__(self,title,author,pages):
      #assignment
      self.title = title
      self.author = author
      self.pages = pages
  
  def __str__(self):
      return f"{self.title} by {self.author}"
  
  def __len__(self):
    return self.pages

  def __del__(self):
    print("An Book object has been deleted.")
```

```
b = Book("Py4e","Jose",250)
```

```
print(b)
```
