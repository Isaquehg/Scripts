```
#Dictionaries can store str, int, float, lists and
#even another dict inside it

prices = {"apple":2.99, "orange":1.98, "lemon":1.56}
print("apple: ", prices["apple"])

d = {"k1": 123, "k2": [1, 2, 3], "k3":{"sub_key":100}}

print(d["k1"])
print(d["k2"][0])
print(d["k3"]["sub_key"])

#Adding key, values
d["k4"] = 300
print(d)

#Dicts METHODS: .keys(), .items(), .items()->makes tuples!
print(d.keys())
print(d.values())
print(d.items())
```
