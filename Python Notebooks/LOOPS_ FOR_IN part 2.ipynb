```
#USING BOOLEAN TO SEARCH.
#E.g.: STIRNGS(abc), INTEGER(1,2,3,4), FLOATING-POINT(.) AND BOOLEAN(T/F)
number = False
print("Before All, number:", number)
for value in (23, 35, 39, 58, 102):
  if value > 40:
    number = True
  print(number, value)
print("After All:", number)

#Once the value of NUMBER was changed, it won´t return!
```

```
#FINFING THE SMALLEST VALUE.
#We haven´t used a random value to SMALLEST. Instead, we´ve used NONE.
smallest = None
for value in (3, 14, 56, 25, 1, 10):
  if smallest is None:
    smallest = value
  elif smallest > value:
    smallest = value
  print(smallest, value)
print("After it, smallest is:", smallest)
```

```
#FINDING THE BIGGEST VALUE

largest = -1
print("Before", largest)
for number in (9, 15, 35, -118, 47, 102, -19, 75, 1013):
    if number > largest:
      largest = number
    print(largest, number)
print("After That:", largest)

#The variable LARGEST tracks variable NUMBER when NUMBER is bigger than the current LARGEST value.
```
