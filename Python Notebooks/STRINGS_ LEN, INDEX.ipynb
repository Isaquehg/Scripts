```
#LEN is teh lenght of a string.

fruit = "Car"
x = len(fruit)
print(x)
```

```
#INDEX
fruit = "Banana"
index = 0
while index < len(fruit):
  letter = fruit[index]
  print(index, letter)
  index = index + 1
print("Finished!")
```

```
#Instead of using index, we can use for/in
fruit = "Banana"
x = len(fruit)
for letter in (fruit):
  print(letter)
print(x, "Letters.")
```

```
#Selecting parts of a string.
#Examples
s = "Monty Python"
print(s[0:4])
print(s[3:7])
print(s[5:14])
print(s[:2])
print(s[6:])
print(s[:])
```
