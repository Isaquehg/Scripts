```
def even_check(number):
  return number % 2 == 0

even_check(20)
```

```
#If there´s just a even number, the result is True!
def even_check_list(list):
  for num in list:
    if num % 2 == 0:
      return True
    else:
      pass
even_check_list([1,2,3])
```
