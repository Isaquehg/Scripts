```
#We can add an ELSE statement when TRY works!If Try don´t works, Else doesn´t works too.

try:
  10 + "10"
except:
  ("Something went wrong.")
else:
  print("Try worked!")
```

```
#Look I can specify the excepts for an exact error.
#It will only run when the specific error occurs.
#finally always occurs

try:
  x = open("testfile","w")
  x.write("testlinee.")
except TypeError:
  print("There was a type error")
except OSError:
  print("There was a OS error")
finally:
  print("I always run!!")
```

```
def ask_for_int():
  try:
    x = int(input("Insert a number: "))
  except:
    print("Ops, this is not a number")
  finally:
    print("End of the operation.")

ask_for_int()
```
