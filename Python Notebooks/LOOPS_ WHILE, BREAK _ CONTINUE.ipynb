```
#LOOP
n = 5
while n > 0:
  print (n)
  n = n - 1
print("Blastoff!")
print(n)
```

```
#LOOPS W/ BREAK.(Go out of the Loop)
while True:
  line = input("What`s next?")
  if line == ("done"):
    break
  print(line)
print("Finished")
```

```
#LOOPS W/ CONTINUE.(Go to the Top of the Loop)
while True:
  line = input("What`s next?")
  if line == "#":
    continue
  elif line == "done":
    break
  else:
    continue
  print(line)
print("Finished.")
```
