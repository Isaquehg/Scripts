```
#Concatenating lists:
a = [1, 2, 3]
b = [4, 5, 6, 7]
c = a + b
print(c)
```

```
#Slicing lists:
a = [12, 45, 28, 69]
print(a[1:3])
print(a[:3])
print(a[2:])
```

```
some = [12, 45, 28, 69]
print("max is:", max(some))
print("min is:", min(some))
print("Len is", len(some))
print("Sum is:", sum(some))
print("Sum / Len is:", sum(some)/len(some))
print("Max + Min is:", max(some)+min(some))
```

```
#Adding things to the list using APPEND function:
stuff = list()
stuff.append("Garry")
stuff.append("Darry")
print(stuff)
stuff.remove("Darry")
print(stuff)
```

```
#Checking if something is in a list:
something = [21, 2, 34, 28]
x = input("Enter a number:")
if x == something[0] or something[1] or something[2] or something[3]:
  print(x, "is in the list!")
else:
  print(x, "is not in the list.")
```

```
#SORT function:
#It removes the list order in strings(or creates it with numbers!).
something = [21, 2, 34, 28]
something.sort()
print(something)
print(something[0])
```

```
#Avarege Counter Using Algorithms:
total = 0
count = 0
while True:
  inp = input("Enter a number:")
  if inp == "done": break
  value = float(inp)
  total = total + value
  count = count + 1
average = total / count
print("The average is", average)
```

```
#Avarege Counter Using Data Structures:
blanklist = list()
while True:
  x = input("Enter a number: ")
  if x == "done": break
  value = float(x)
  blanklist.append(value)
avarege = sum(blanklist) / len(blanklist)
print("The avarege is:", avarege)
```
