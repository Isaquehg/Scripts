```
#MAP(func,paramether)
#MAP applies a FUNCTION into a SPECIFIC element outside the function.
#The map output can be a for loop or a list.

def square(num):
  return num**2

mylist = [1,2,3,4,5]
```

```
for item in map(square,mylist):
  print(item)
```

```
list(map(square,mylist))
```

```
#FILTER
#Filter does the same as map, but aplies just for True results.
mynums = [1,2,3,4,5,6,7]
def check_even(num):
  return num % 2 == 0
```

```
list(filter(check_even,mynums))
```

```
for item in filter(check_even,mynums):
  print(item)
```

```
#LAMBDA
#name_of_function = lambda item: (functon content,statements,etc.)
#Note that normally we don´t use the function name before Lambda...

#def square(num):
#  result = num**2
#  return result

#lambda version of that function:
square = lambda num = num**2
```

```
#applying map in lambda expressions:
list(map(lambda num: num**2,mynums))
```

```
#applying filter to lambda expressions.
list(filter(lambda num: num%2 == 0,mynums))
```
